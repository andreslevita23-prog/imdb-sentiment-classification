## Marco de evaluación

Para comparar modelos con un criterio homogéneo, se utiliza una rutina de evaluación común. Además del F1, se revisan ROC AUC, Average Precision y el comportamiento por umbral, ya que en una aplicación de moderación importa tanto detectar reseñas negativas como controlar falsos positivos.

```python
import sklearn.metrics as metrics
def evaluate_model(model, train_features, train_target, test_features, test_target):

    eval_stats = {}

    fig, axs = plt.subplots(1, 3, figsize=(20, 6))

    for type, features, target in (('train', train_features, train_target), ('test', test_features, test_target)):

        eval_stats[type] = {}

        pred_target = model.predict(features)
        pred_proba = model.predict_proba(features)[:, 1]

        # F1
        f1_thresholds = np.arange(0, 1.01, 0.05)
        f1_scores = [metrics.f1_score(target, pred_proba>=threshold) for threshold in f1_thresholds]

        # ROC
        fpr, tpr, roc_thresholds = metrics.roc_curve(target, pred_proba)
        roc_auc = metrics.roc_auc_score(target, pred_proba)
        eval_stats[type]['ROC AUC'] = roc_auc

        # PRC
        precision, recall, pr_thresholds = metrics.precision_recall_curve(target, pred_proba)
        aps = metrics.average_precision_score(target, pred_proba)
        eval_stats[type]['APS'] = aps

        if type == 'train':
            color = 'blue'
        else:
            color = 'green'

        # Valor F1
        ax = axs[0]
        max_f1_score_idx = np.argmax(f1_scores)
        ax.plot(f1_thresholds, f1_scores, color=color, label=f'{type}, max={f1_scores[max_f1_score_idx]:.2f} @ {f1_thresholds[max_f1_score_idx]:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(f1_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(f1_thresholds[closest_value_idx], f1_scores[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('threshold')
        ax.set_ylabel('F1')
        ax.legend(loc='lower center')
        ax.set_title(f'Valor F1')

        # ROC
        ax = axs[1]
        ax.plot(fpr, tpr, color=color, label=f'{type}, ROC AUC={roc_auc:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(roc_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(fpr[closest_value_idx], tpr[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.plot([0, 1], [0, 1], color='grey', linestyle='--')
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('FPR')
        ax.set_ylabel('TPR')
        ax.legend(loc='lower center')
        ax.set_title(f'Curva ROC')

        # PRC
        ax = axs[2]
        ax.plot(recall, precision, color=color, label=f'{type}, AP={aps:.2f}')
        # establecer cruces para algunos umbrales
        for threshold in (0.2, 0.4, 0.5, 0.6, 0.8):
            closest_value_idx = np.argmin(np.abs(pr_thresholds-threshold))
            marker_color = 'orange' if threshold != 0.5 else 'red'
            ax.plot(recall[closest_value_idx], precision[closest_value_idx], color=marker_color, marker='X', markersize=7)
        ax.set_xlim([-0.02, 1.02])
        ax.set_ylim([-0.02, 1.02])
        ax.set_xlabel('recall')
        ax.set_ylabel('precision')
        ax.legend(loc='lower center')
        ax.set_title(f'PRC')

        eval_stats[type]['Accuracy'] = metrics.accuracy_score(target, pred_target)
        eval_stats[type]['F1'] = metrics.f1_score(target, pred_target)

    df_eval_stats = pd.DataFrame(eval_stats)
    df_eval_stats = df_eval_stats.round(2)
    df_eval_stats = df_eval_stats.reindex(index=('Accuracy', 'F1', 'APS', 'ROC AUC'))

    print(df_eval_stats)

    return
```

## Preprocesamiento de texto

Se crea una versión normalizada del texto para reducir ruido superficial y asegurar que los modelos comparen señales lingüísticas consistentes en lugar de variaciones de formato.

```python
import re

def basic_text_clean(text):
    # pasar a minúsculas
    text = str(text).lower()
    # eliminar todo lo que no sea letra o espacio
    text = re.sub(r'[^a-z\s]', ' ', text)
    # colapsar espacios multiples
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_reviews['review_norm'] = df_reviews['review'].apply(basic_text_clean)
```

El dataset ya incluye una separación explícita entre entrenamiento y prueba mediante la columna `ds_part`. Se respeta esta partición para preservar una evaluación comparable y evitar fugas de información.

## Modelos

### Benchmark de negocio: clasificador de mayoría

Antes de evaluar modelos de NLP, se establece un punto de referencia mínimo: un clasificador que siempre predice la clase mayoritaria. Si un modelo no supera este benchmark con claridad, no aporta valor operativo.

```python
from sklearn.dummy import DummyClassifier
```

```python
df_train = df_reviews[df_reviews['ds_part'] == 'train']
df_test = df_reviews[df_reviews['ds_part'] == 'test']

X_train = df_train['review_norm']
y_train = df_train['pos']

X_test = df_test['review_norm']
y_test = df_test['pos']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1,2),
    stop_words='english'
)

# Ajustar el vectorizador únicamente con entrenamiento
tfidf.fit(df_train['review_norm'])

# Transformar entrenamiento y prueba con el mismo vocabulario
X_train_tfidf = tfidf.transform(df_train['review_norm'])
X_test_tfidf = tfidf.transform(df_test['review_norm'])
```

```python
# Benchmark de negocio: siempre predecir la clase mayoritaria

dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(X_train_tfidf, y_train)


print("BUSINESS BENCHMARK - MAJORITY CLASSIFIER")
evaluate_model(dummy, X_train_tfidf, y_train, X_test_tfidf, y_test)

```

### Candidato 1: TF-IDF + Logistic Regression

Este primer modelo combina una representación TF-IDF con regresión logística balanceada. Es una línea base fuerte para clasificación de texto porque ofrece buen rendimiento, entrenamiento rápido y una implementación sencilla de mantener.

```python
from sklearn.linear_model import LogisticRegression

```

```python
# Candidato 1: Logistic Regression sobre representaciones TF-IDF
lr = LogisticRegression(
    max_iter=2000,
    n_jobs=-1,
    class_weight='balanced'
)

lr.fit(X_train_tfidf, y_train)

print("CANDIDATE 1 - TF-IDF + LOGISTIC REGRESSION")
evaluate_model(lr, X_train_tfidf, y_train, X_test_tfidf, y_test)
```

### Candidato 2: TF-IDF + Calibrated Linear SVC

Este candidato mantiene la misma representación TF-IDF, pero reemplaza la regresión logística por un Linear SVC calibrado para obtener probabilidades utilizables en las curvas de evaluación.

```python
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV

base_svc = LinearSVC()
svc_calibrated = CalibratedClassifierCV(base_svc, cv=3)

svc_calibrated.fit(X_train_tfidf, y_train)

print("CANDIDATE 2 - TF-IDF + CALIBRATED LINEAR SVC")
evaluate_model(svc_calibrated, X_train_tfidf, y_train, X_test_tfidf, y_test)


```

### Candidato 3: spaCy Lemmatization + TF-IDF + Logistic Regression

El tercer enfoque añade lematización con spaCy antes de vectorizar el texto. La hipótesis es que una normalización lingüística más profunda podría mejorar la generalización del modelo sin cambiar el clasificador final.

```python
import spacy

nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
def text_preprocessing_3(text):

    doc = nlp(text)
    tokens = [token.lemma_ for token in doc]

    return ' '.join(tokens)
```

```python
df_reviews['review_spacy'] = df_reviews['review'].apply(text_preprocessing_3)
```

```python
df_train_spacy = df_reviews[df_reviews['ds_part'] == 'train']
df_test_spacy  = df_reviews[df_reviews['ds_part'] == 'test']

X_train_spacy = df_train_spacy['review_spacy']
y_train_spacy = df_train_spacy['pos']

X_test_spacy = df_test_spacy['review_spacy']
y_test_spacy = df_test_spacy['pos']
```

```python
tfidf_spacy = TfidfVectorizer(
    max_features=50000,
    ngram_range=(1,2),
    stop_words='english'
)

tfidf_spacy.fit(X_train_spacy)

X_train_spacy_tfidf = tfidf_spacy.transform(X_train_spacy)
X_test_spacy_tfidf  = tfidf_spacy.transform(X_test_spacy)
```

```python
lr_spacy = LogisticRegression(
    max_iter=2000,
    n_jobs=-1,
    class_weight='balanced'
)

lr_spacy.fit(X_train_spacy_tfidf, y_train_spacy)

print("CANDIDATE 3 - SPACY LEMMATIZATION + TF-IDF + LOGISTIC REGRESSION")
evaluate_model(
    lr_spacy,
    X_train_spacy_tfidf, y_train_spacy,
    X_test_spacy_tfidf, y_test_spacy
)
```

## Comparación de modelos

Los tres candidatos superan ampliamente el benchmark de negocio, que obtiene un F1 de 0.00 y métricas equivalentes al azar. A partir de ahí, la comparación relevante es entre rendimiento, estabilidad y costo de implementación.

- **Candidato 1: TF-IDF + Logistic Regression** alcanza un F1 de 0.88 en prueba con una configuración simple y fácil de explicar.
- **Candidato 2: TF-IDF + Calibrated Linear SVC** también logra un F1 de 0.88, aunque con un ajuste más agresivo en entrenamiento (F1 de 0.99), lo que sugiere un riesgo mayor de sobreajuste.
- **Candidato 3: spaCy Lemmatization + TF-IDF + Logistic Regression** vuelve a obtener un F1 de 0.88, pero añade una capa extra de procesamiento sin mejorar el resultado final.

En términos prácticos, el dataset responde muy bien a enfoques clásicos de NLP. Las curvas ROC y PRC confirman un comportamiento estable en prueba, y la similitud de resultados sugiere que el mayor valor proviene de una representación TF-IDF sólida más que de un pipeline lingüístico más complejo.

## Recomendación final

La mejor opción para llevar a una primera versión productiva es **TF-IDF + Logistic Regression**. Iguala el mejor F1 observado en prueba (0.88), supera con claridad el umbral objetivo de 0.85 y ofrece la mejor relación entre desempeño, simplicidad operativa y facilidad de mantenimiento.

El modelo calibrado con Linear SVC es competitivo, pero no aporta una ganancia tangible sobre la regresión logística. La variante con lematización en spaCy confirma que un preprocesamiento más costoso no mejora el resultado en este dataset.

Como siguiente paso de negocio, este modelo puede integrarse como filtro de priorización para moderación o análisis de satisfacción. Si el caso de uso exigiera mayor recall sobre reseñas negativas, el umbral de decisión podría ajustarse con apoyo de las curvas F1, ROC y PRC ya calculadas en este notebook.
